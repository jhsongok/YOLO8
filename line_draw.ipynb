## Polylines Draw

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

img = cv2.imread("/content/test2.png")

pts = np.array([[200, 500], [1050, 500], [1050, 460], [200, 460]], np.int32)
pts = [pts]

cv2.polylines(img, pts, isClosed=True, color=(0, 255, 0), thickness=3)

cv2_imshow(img)
```

## Line Draw

```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/speed_video.mp4_000005630.png")

cv2.line(img, (0, 400), (1280, 400), (0, 255, 0), thickness=4)

cv2_imshow(img)
```
